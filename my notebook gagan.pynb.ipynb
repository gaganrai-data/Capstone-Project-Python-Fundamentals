```python
# 📦 Import necessary libraries
import pandas as pd
import numpy as np

# 🧱 Define Project DataFrame
project_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Project Cost': [100000, 150000, np.nan, 120000, 130000],
    'Status': ['Ongoing', 'Failed', 'Finished', 'Ongoing', 'Failed']
}
df_project = pd.DataFrame(project_data)

# 👥 Define Employee DataFrame
employee_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['Mr. Luxume', 'Mrs. Rina Roy', 'Mr. John Smith', 'Mrs. Alia Khan', 'Mr. Steve'],
    'Age': [34, 28, 45, 31, 26],
    'Gender': ['F', 'F', 'M', 'F', 'M'],
    'City': ['London', 'Delhi', 'New York', 'Tokyo', 'Oslo']
}
df_employee = pd.DataFrame(employee_data)

# 🏷️ Define Seniority Level DataFrame
seniority_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Designation Level': [1, 2, 3, 4, 5]
}
df_seniority = pd.DataFrame(seniority_data)

# 💾 Save all three DataFrames as CSV files
df_project.to_csv('project.csv', index=False)
df_employee.to_csv('employee.csv', index=False)
df_seniority.to_csv('seniority.csv', index=False)

# ✅ Confirm creation
print("CSV files created: project.csv, employee.csv, seniority.csv")
```

```python
# 📥 Load the Project DataFrame from CSV
df_project = pd.read_csv('project.csv')

# 🧮 Replace missing values in 'Project Cost' using running average
running_sum = 0
count = 0

for i in range(len(df_project)):
    if pd.isna(df_project.loc[i, 'Project Cost']):
        avg = running_sum / count if count > 0 else 0
        df_project.loc[i, 'Project Cost'] = avg
    else:
        running_sum += df_project.loc[i, 'Project Cost']
        count += 1

# 💾 Save the updated DataFrame
df_project.to_csv('project.csv', index=False)

# ✅ Preview result
print(df_project)
```

```python
# 📥 Load employee data
df_employee = pd.read_csv('employee.csv')

# ✂️ Split 'Name' into 'First Name' and 'Last Name'
# This assumes names are in the format: "Mr./Mrs. First Last"
df_employee[['First Name', 'Last Name']] = df_employee['Name'].str.split(' ', n=1, expand=True)

# 🧹 Drop the original 'Name' column
df_employee.drop(columns=['Name'], inplace=True)

# 💾 Save updated employee data
df_employee.to_csv('employee.csv', index=False)

# ✅ Preview result
print(df_employee)
```

```python
# 📥 Load all three CSVs
df_project = pd.read_csv('project.csv')
df_employee = pd.read_csv('employee.csv')
df_seniority = pd.read_csv('seniority.csv')

# 🔗 Merge employee and seniority data
df_merged = pd.merge(df_employee, df_seniority, on='ID', how='inner')

# 🔗 Merge with project data
df_final = pd.merge(df_merged, df_project, on='ID', how='inner')

# 💾 Save final merged DataFrame
df_final.to_csv('final_merged.csv', index=False)

# ✅ Preview result
print(df_final)
```

```python
# 📥 Load the final merged DataFrame
df_final = pd.read_csv('final_merged.csv')

# 💰 Calculate 5% bonus based on Project Cost
df_final['Bonus'] = df_final['Project Cost'] * 0.05

# 🎯 Mark who received a bonus (non-zero)
df_final['Received Bonus'] = df_final['Bonus'] > 0

# 💾 Save the updated DataFrame
df_final.to_csv('final_merged.csv', index=False)

# ✅ Preview bonus info
print(df_final[['ID', 'First Name', 'Project Cost', 'Bonus', 'Received Bonus']])
```

```python
# 📥 Load the final merged DataFrame
df_final = pd.read_csv('final_merged.csv')

# 🔁 Loop through rows to demote designation level for failed projects
for i in range(len(df_final)):
    if df_final.loc[i, 'Status'].lower() == 'failed':
        current_level = df_final.loc[i, 'Designation Level']
        # Demote by 1, but don't exceed level 5
        df_final.loc[i, 'Designation Level'] = min(current_level + 1, 5)

# 🧹 Remove records with designation level above 4
df_final = df_final[df_final['Designation Level'] <= 4]

# 💾 Save the updated DataFrame
df_final.to_csv('final_merged.csv', index=False)

# ✅ Preview updated designation levels
print(df_final[['ID', 'First Name', 'Status', 'Designation Level']])
```

```python
# 📥 Load the final merged DataFrame
df_final = pd.read_csv('final_merged.csv')

# 🧼 Remove 'Mr.' and 'Mrs.' from the First Name column
df_final['First Name'] = df_final['First Name'].str.replace('Mr.', '', regex=False)
df_final['First Name'] = df_final['First Name'].str.replace('Mrs.', '', regex=False)

# 🧹 Drop the Gender column
df_final.drop(columns=['Gender'], inplace=True)

# 💾 Save the updated DataFrame
df_final.to_csv('final_merged.csv', index=False)

# ✅ Preview cleaned names
print(df_final[['ID', 'First Name', 'Last Name', 'City']])
```

```python
# 📥 Load the final merged DataFrame
df_final = pd.read_csv('final_merged.csv')

# 🔁 Loop through each row to promote designation level
for i in range(len(df_final)):
    if df_final.loc[i, 'Age'] > 29:
        current_level = df_final.loc[i, 'Designation Level']
        # Promote by 1, but don't go above level 1
        df_final.loc[i, 'Designation Level'] = max(current_level - 1, 1)

# 💾 Save the updated DataFrame
df_final.to_csv('final_merged.csv', index=False)

# ✅ Preview updated levels
print(df_final[['ID', 'First Name', 'Age', 'Designation Level']])
```

```python
# 📥 Load the final merged DataFrame
df_final = pd.read_csv('final_merged.csv')

# 📊 Group by ID and First Name, then sum the Project Cost
df_total = df_final.groupby(['ID', 'First Name'])['Project Cost'].sum().reset_index()

# 🏷️ Rename the column to 'Total Cost'
df_total.rename(columns={'Project Cost': 'Total Cost'}, inplace=True)

# 💾 Save the new DataFrame as TotalProjCost
df_total.to_csv('TotalProjCost.csv', index=False)

# ✅ Preview the result
print(df_total)
```
